```python
# Data link
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data"
```

```python
names = ['Sex','Length','Diameter','Height','Whole weight','Shucked weight','Viscera weight','Shell weight','Rings']
data = pd.read_csv(url, header=None, names = names)
data.head()
```

```python
replace_list = {"Sex" : {"M": 0, "F" : 1, "I": 2}}
data.replace(replace_list,inplace=True)
```

```python
#data.head()
#data.tail()
```

```python
# x and y defining
X = data.iloc[:,:8].values
y = data.iloc[:,8].values
```

```python
# Data set splitting into a training and a test set
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
# Parameters 
param_grid = [{'regression': [Ridge()], 
         'regression__alpha': [0.001, 0.01, 0.1, 1, 10, 100]},
        {'regression': [KNeighborsRegressor()],
         'regression__n_neighbors': [1,4,5,6,8,10,15,20,100]}]
```

```python
# Pipeline
pipe = Pipeline([('preprocessing', StandardScaler()), ('regression', Ridge())])

# Tuning (GridSearch)
grid = GridSearchCV(pipe, param_grid=param_grid, cv=5, n_jobs=-1)

# Fitting 
grid.fit(X_train, y_train)
mean_ridge=grid.cv_results_['mean_test_score']
```

```python
print("Best: %f using %s" % (grid.best_score_, grid.best_params_))

print("Best parameters: \n{}".format(grid.best_params_))

print("Best cross-validation score: {:.2f}".format(grid.best_score_))

print("Test set score: {:.2f}".format(grid.score(X_test, y_test)))
```

```python
mean_a = grid.cv_results_['mean_test_score']
std_a = grid.cv_results_['std_test_score']
param_a = grid.cv_results_['params']
for mean, stdev, param in zip(mean_a, std_a, param_a):
    print("%f (%f) with: %r" % (mean, stdev, param))
```

```python
print("The best estimator:\n{}".format(grid.best_estimator_))
```
